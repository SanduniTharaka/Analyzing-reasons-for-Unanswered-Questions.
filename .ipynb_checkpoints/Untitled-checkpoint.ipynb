```python
from sklearn.preprocessing import MinMaxScaler # for scaling data
```

```python
#Scaling the dataset
mms=MinMaxScaler()
mms.fit(X)
data_transformed = mms.transform(X) 
```

```python
data_transformed
```

```python
#Elbow method to minimize (Within-cluster sum of square)
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
#Plotting the elbow curve by k (no. of clusters)
plt.plot(K,Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squred_distances')
plt.title('Elbow Method For optimal k')
plt.show()
```

```python
%%time
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=2).fit_predict(data_transformed)
print(silhouette_score(data_transformed, kmeans))
```
