```python
#Elbow method to minimize (Within-cluster sum of square)
Sum_of_squared_distances = []
K = range(1,5)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
#Plotting the elbow curve by k (no. of clusters)
plt.plot(K,Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squred_distances')
plt.title('Elbow Method For optimal k')
plt.show()
```

```python
%%time
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=2).fit_predict(X)
print(silhouette_score(X, kmeans))
```
